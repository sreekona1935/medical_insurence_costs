The BMI distribution indicates a population skewed toward overweight and obese categories, with substantial variability and a small number of extreme high values. BMI is well-populated and suitable for modeling, though its skew suggests potential benefit from categorization or interaction analysis.

```python
df['charges'].describe()
#mean > median so it is right skew
```

Medical insurance charges exhibit substantial right skew, with a small proportion of individuals accounting for disproportionately high costs, resulting in a mean that overstates the typical expenditure.
