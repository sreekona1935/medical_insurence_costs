```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['charges'], bins=30)
plt.title("Distribution of Medical Insurance Charges")
plt.show()
```

```python
sns.boxplot(x='smoker', y='charges', data=df)
plt.title("Insurance Charges by Smoking Status")
plt.show()
```

```python
sns.scatterplot(x='age', y='charges', data=df)
plt.title("Age vs Insurance Charges")
plt.show()
```

```python
sns.scatterplot(x='age', y='charges', hue='smoker', data=df)
plt.title("Age vs Charges (Colored by Smoking Status)")
plt.show()
```

```python
# Define BMI bins and labels
bmi_bins = [0, 24, 29, 34, 39, df['bmi'].max()]
bmi_labels = ['024', '2529', '3034', '3539', '39+']

# Create BMI category column
df['bmi_range'] = pd.cut(df['bmi'], bins=bmi_bins, labels=bmi_labels)
#df['bmi_range'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(
    x='bmi_range',
    y='charges',
    #hue= 'smoker',
    data=df,
    errorbar=None
)

plt.title("Average Medical Charges by BMI Range")
plt.xlabel("BMI Range")
plt.ylabel("Average Insurance Charges")
plt.show()
```

## Initial Observations 

#### Age, Smoking, and charges
As age increases, insurense charges genearlly tend to increase. Older individuals are more likely to incur higher medical costs. And though belonging to the same age group people who are smokers have higher charges compared to non-smokers as smoking depletes health.

#### BMI's effect on charges
The charges are also increasisng the higher the BMI is. So higher the body fat percentage the more likely health isuues
